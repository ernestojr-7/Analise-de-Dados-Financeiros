## Base de dados com mais ações

- [BOVA11](https://www.sunoresearch.com.br/artigos/bova11/)

```
acoes = ['GOLL4.SA', 'CVCB3.SA', 'WEGE3.SA', 'MGLU3.SA', 'TOTS3.SA', 'BOVA11.SA', 'ABEV3.SA', 'ODPV3.SA', 'VIVT3.SA', 'PETR3.SA', 'BBAS3.SA']
```

```
df_acoes = pd.DataFrame()
for acao in acoes:
  df_acoes[acao] = yf.download(acao, start='2015-01-01')['Close']
```

```
df_acoes
```

```
df_acoes = df_acoes.rename(columns={'GOLL4.SA': 'GOL', 'CVCB3.SA': 'CVC', 'WEGE3.SA': 'WEGE',
                                    'MGLU3.SA': 'MGLU', 'TOTS3.SA': 'TOTS', 'BOVA11.SA': 'BOVA',
                                    'ABEV3.SA': 'AMBEV', 'ODPV3.SA': 'ODONTOPREV', 'VIVT3.SA': 'VIVO',
                                    'PETR3.SA': 'PETROBRAS', 'BBAS3.SA': 'BBRASIL'})
```

Supondo que as ações foram adquiridas em 2015-01-02, os dados foram normalizados em relação ao valor das ações no dia da aquisição.



```
df_acoes_normalizado = df_acoes.copy()
for i in df_acoes_normalizado.columns[1:]:
  df_acoes_normalizado[i] = df_acoes_normalizado[i] / df_acoes_normalizado[i][0]
```

```
df_acoes_normalizado
```

```
df_acoes_normalizado.plot(x = 'Date', figsize = (15,7), title = 'Histórico do preço das ações - normalizado');
```

```
figura = px.line(title = 'Histórico do preço das ações')
for i in df_acoes.columns[1:]:
  figura.add_scatter(x = df_acoes['Date'], y = df_acoes[i], name = i)
figura.show()
```

```
figura = px.line(title = 'Histórico do preço das ações - normalizado')
for i in df_acoes_normalizado.columns[1:]:
  figura.add_scatter(x = df_acoes_normalizado['Date'], y = df_acoes_normalizado[i], name = i)
figura.show()
```
